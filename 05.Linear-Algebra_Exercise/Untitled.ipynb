```python
from numpy.polynomial import Polynomial
import numpy as np
import matplotlib.pyplot as plt
import sympy

from PIL import Image
import colorsys
import math
import os
import itertools
from sympy.ntheory import factorint
import time
import timeit
import random
from functools import partial
import secrets
```

```python
x = [50, 20, 55, 30, 32, 33, 35, 250, 280]
y = [secrets.randbits(k) for k in x]

factors = [factorint(p) for p in y]
numbers = []
print(factors)
start_time = timeit.timeit()
for factor in factors:
    for num, times in factor.items():
        numbers.append(num * times)
end_time = timeit.timeit()
print(start_time)
print(end_time)
print(end_time - start_time)
duration= end_time - start_time

print(duration)
```
