```python
import numpy as np
import matplotlib.pyplot as plt

# Sampling rate and duration
fs = 1000  # Sampling rate (samples per second)
t = np.linspace(0, 1, fs)  # One second worth of samples

# Frequencies of the sine waves
frequencies = [50, 150, 300]  # Hz

# Generate sine waves and sum them to create a complex signal
signal = sum(np.sin(2 * np.pi * f * t) for f in frequencies)

# Plot the individual sine waves
plt.figure(figsize=(12, 10))
for i, f in enumerate(frequencies):
    plt.subplot(4, 1, i + 1)
    plt.plot(t, np.sin(2 * np.pi * f * t))
    plt.title(f'Sine wave with frequency {f} Hz')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')

# Plot the combined signal
plt.subplot(4, 1, 4)
plt.plot(t, signal)
plt.title('Combined signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

```

```python
# Compute the Fast Fourier Transform (FFT) of the combined signal
fft_signal = np.fft.fft(signal)
fft_freqs = np.fft.fftfreq(len(signal), 1/fs)

# Plot the magnitude of the FFT (only the positive frequencies)
plt.figure(figsize=(10, 4))
plt.plot(fft_freqs[:len(fft_freqs)//2], np.abs(fft_signal)[:len(fft_signal)//2])
plt.title('Frequency domain representation')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
